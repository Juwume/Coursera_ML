```python
import pandas as pd
import numpy as np

from scipy import linalg
import math
```

```python
def study_func(x):
    return math.sin(x / 5.) * math.exp(x / 10.) + 5. * math.exp(-x / 2.)
def create_A(matrix_x):
    out = np.zeros((len(matrix_x),len(matrix_x)))
    for row in range(len(matrix_x)):
        for col in range(len(matrix_x)):
            out[row][col] = math.pow(matrix_x[col],row)
    return out
```

```python
A = create_A([1,15])
print(A)
```

```python

b = np.array(list(map(study_func,[1,15])))
print(b)
```

```python
x = linalg.solve(A,b)
print(x)
```
