```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as sts
%matplotlib inline
```

```python
k = 2
gamma = sts.gamma(k)
sample = gamma.rvs(1000)
sample[:10]
```

## Теоретическая и эмпирическая функции распределения

```python
x = np.linspace(0,10,100)
cdf = gamma.cdf(x)
plt.plot(x,cdf, label='theoretical CDF')

from statsmodels.distributions.empirical_distribution import ECDF
ecdf = ECDF(sample)
plt.step(ecdf.x, ecdf.y, label='ECDF')

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='lower right')

plt.show()
```

## Теоретическая и эмпирическая функции плотности

```python
plt.plot(x, gamma.pdf(x), label='Theoretical')
plt.hist(sample, density=True, label='Empirical')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper right')
plt.show()
```

## Величины гамма распределения
В данном случае теоретическая средняя равна k=2.
Дисперсия также равна k=2

## Возьмём 1000 выборок размером 5
### Таким образом выборочная средняя должна быть распределена:
$$ X ≈∼ N(EX, \frac{DX}{n})$$

### Для n = 5 дисперсия равна $\frac{2}{5}$ 
#### Следовательно сигма равна $\sqrt{\frac{2}{5}}$

```python
normal = sts.norm(loc=2, scale=(2./5)**0.5)
x_norm = np.linspace(0,4,100)
plt.plot(x_norm, normal.pdf(x_norm), label='Theoretical')
bars = []
temp = gamma.rvs(size=(1000,5))
for i in range(1000):
    bars.append(temp[i].mean())
    
plt.hist(bars, bins=25, density=True, label='Empirical')

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper right')
plt.show()
```
